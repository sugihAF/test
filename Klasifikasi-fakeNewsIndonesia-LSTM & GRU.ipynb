---
FEATURE ENGINEERING

```python
df['isfake'] = df.tagging.map({'Valid':0,'Hoax':1})
df
```

---
STEMMING

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
hasil_stem = []
for kalimat in df['berita']:
    stem = stemmer.stem(kalimat)
    hasil_stem.append(stem)
```

```python
df['stemming'] = hasil_stem
```

```python
df['isfake'].value_counts()
```

---
TEXT PREPROCESSING

```python
#filter STOPWORD
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('indonesian')
```

```python
stop_words
```

```python
import gensim
# Remove stopwords and remove words with 2 or less characters
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3 and token not in stop_words:
            result.append(token)
            
    return result
```

```python
df['filtered'] = df['stemming'].apply(preprocess)
```

```python
df
```

```python
print(df['filtered'][59])
```

```python
maxlen = 1
for word in df.filtered:
    if maxlen<len(word):
        maxlen = len(word)
maxlen
```

```python
df['filtered_join'] = df['filtered'].apply(lambda x: " ".join(x))
```

```python
print(df['filtered_join'][0])
```

---
SPLIT DATA

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.filtered_join, df.isfake, test_size = 0.2)
```

---
TOKENIZING

```python
# Obtain the total words present in the dataset
list_of_words = []
for i in df.filtered:
    for j in i:
        list_of_words.append(j)
```

```python
len(list_of_words)
```

```python
#ambil kata unik dengan set
total_words = len(list(set(list_of_words)))
total_words
```

```python
from nltk import word_tokenize
from tensorflow.keras.preprocessing.text import one_hot, Tokenizer
```

```python
tokenizer = Tokenizer(num_words = total_words)
tokenizer.fit_on_texts(x_train)
train_sequences = tokenizer.texts_to_sequences(x_train)
test_sequences = tokenizer.texts_to_sequences(x_test)
```

```python
print("Asli :\n",df.filtered_join[59],"\n Tokenized : ",train_sequences[59])
```

---
ADD PADDING

```python
df['len'] = df['filtered'].apply(lambda x: len(x))
```

```python
df['len'].mean()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_train = pad_sequences(train_sequences,maxlen = 163, padding = 'post', truncating = 'post')
padded_test = pad_sequences(test_sequences,maxlen = 163, truncating = 'post') 
```

---
MAKE MODEL

```python
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Embedding, Input, LSTM, Conv1D, MaxPool1D, Bidirectional, GRU
from tensorflow.keras.models import Model
```

```python
y_train = np.asarray(y_train)
y_test = np.asarray(y_test)
```

---
LSTM

```python
from keras import layers
from keras.models import Sequential #used to initialize our ANN
from keras.layers import Dense #used to make layers in ANN
# Sequential Model
model_lstm = Sequential()
# embeddidng layer
model_lstm.add(layers.Embedding(total_words, output_dim = 240))
# Bi-Directional RNN and LSTM
model_lstm.add(layers.LSTM(128))
# Menambah layer RNN
model_lstm.add(layers.Dense(128, activation = 'relu'))
model_lstm.add(layers.Dense(1,activation= 'sigmoid'))
model_lstm.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
model_lstm.summary()
```

```python
#fit data train ke model
model_lstm.fit(padded_train, y_train, batch_size = 64, validation_split = 0.1, epochs = 15)
```

---
GRU

```python
# Sequential Model
model_gru = Sequential()
# embeddidng layer
model_gru.add(layers.Embedding(total_words, output_dim = 240))
# Bi-Directional RNN and LSTM
model_gru.add(layers.GRU(128))
# Menambah layer RNN
model_gru.add(layers.Dense(128, activation = 'relu'))
model_gru.add(layers.Dense(1,activation= 'sigmoid'))
model_gru.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
model_gru.summary()
```

```python
#fit data train ke model
model_gru.fit(padded_train, y_train, batch_size = 64, validation_split = 0.1, epochs = 15)
```

---
HASIL

```python
pred_lstm = model_lstm.predict(padded_test)
pred_gru = model_gru.predict(padded_test)
```

```python
# if the predicted value is >0.5 it is real else it is fake
prediction_lstm = []
for i in range(len(pred_lstm)):
    if pred_lstm[i].item() > 0.5:
        prediction_lstm.append(1)
    else:
        prediction_lstm.append(0)
```

```python
# if the predicted value is >0.5 it is real else it is fake
prediction_gru = []
for i in range(len(pred_gru)):
    if pred_gru[i].item() > 0.5:
        prediction_gru.append(1)
    else:
        prediction_gru.append(0)
```

```python
#scoring dengan matrik accuracy
from sklearn.metrics import accuracy_score
accuracy_lstm = accuracy_score(list(y_test), prediction_lstm)
accuracy_gru = accuracy_score(list(y_test), prediction_gru)
print("Model LSTM Accuracy : ", accuracy_lstm)
print("Model GRU Accuracy : ",accuracy_gru)
```

```python
from sklearn.metrics import roc_auc_score
roc_score_lstm = roc_auc_score(y_test,prediction_lstm)
roc_score_gru = roc_auc_score(y_test,prediction_gru)
print("ROC AUC LSTM SCORE : ",roc_score_lstm)
print("ROC AUC GRU SCORE : ",roc_score_gru)
```

TUNING MODEL

```python
#TUNING OUR ANN
import keras
from keras import layers
from keras.models import Sequential #used to initialize our ANN
from keras.layers import Dense #used to make layers in ANN
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
def build_classifier(optimizer): #we build the classifier using K-FOLD method , so this is basicly same with classifier that we've been made but now we using K-FOLD for the training 
    # Sequential Model
    classifier = Sequential()
    # embeddidng layer
    classifier.add(layers.Embedding(total_words, output_dim = 240))
    # Bi-Directional RNN and LSTM
    classifier.add(layers.LSTM(128))
    # Menambah layer RNN
    classifier.add(layers.Dense(128, activation = 'relu'))
    classifier.add(layers.Dense(1,activation= 'sigmoid'))
    classifier.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['acc'])
    return classifier
classifier = KerasClassifier(build_fn = build_classifier)
parameters = {'batch_size' : [25, 32, 64],
              'epochs' : [10, 15, 20],
              'optimizer' : ['adam', 'rmsprop']}
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 10)
grid_search = grid_search.fit(padded_train, y_train)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
